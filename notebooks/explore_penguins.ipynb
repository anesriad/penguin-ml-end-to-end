```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/riadanas/Desktop/penguin ML end2end/penguins.csv")
df.head()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# basic stats
df.describe(include='all')
```

```python
# histograms of numeric features
df[["culmen_length_mm","culmen_depth_mm","flipper_length_mm","body_mass_g"]].hist(figsize=(8,6))
plt.tight_layout()
plt.show()
```

```python
# Scatter plot
plt.scatter(df["culmen_length_mm"], df["flipper_length_mm"], alpha=0.6)
plt.xlabel("culmen_length_mm")
plt.ylabel("flipper_length_mm")
plt.show()
```

```python
# count of each species
df["species"].value_counts().plot.bar()
plt.ylabel("count")
plt.show()
```

```python
# save cleaned data
df.dropna(subset=["species", 
                  "culmen_length_mm", 
                  "culmen_depth_mm", 
                  "flipper_length_mm", 
                  "body_mass_g"]) \
  .to_csv("/Users/riadanas/Desktop/penguin ML end2end/data/processed/penguins_clean.csv", index=False)
```
